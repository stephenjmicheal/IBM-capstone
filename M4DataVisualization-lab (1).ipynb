# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib as mtl
import matplotlib.pyplot as plt
mtl.style.use(['ggplot'])
```

```python
# your code goes here
Query='''
SELECT ConvertedComp FROM Master
'''
concomp=pd.read_sql(Query,conn)

concomp.plot(kind='hist',color='red',alpha=0.35,figsize=(10,7))
plt.title('Histogram of Converted Compensation')
plt.xlabel('Compensation')
plt.ylabel('Distribution')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
Query='''
SELECT Age FROM Master
'''
Age=pd.read_sql(Query,conn)
Age.plot(kind='box',figsize=(10,7),color='darkblue')
plt.title('Box plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
Query='''
SELECT Age,WorkWeekHrs FROM
Master
'''
Age_Hrs=pd.read_sql(Query,conn)
Age_Hrs.plot(kind='scatter',figsize=(10,7),x='Age',y='WorkWeekHrs',color='darkgreen')
plt.xlabel('Age')
plt.ylabel('Work Week Hrs')
plt.title('Scatter plot of Age and WorkWeekHrs')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
Query='''
SELECT WorkWeekHrs,Age FROM Master
'''
WorkWeekHrs=pd.read_sql(Query,conn)
Query='''
SELECT CodeRevHrs,Age FROM Master
'''
CodeRevHrs=pd.read_sql(Query,conn)

#using Age as the determinant variable
#First, normalized the specified columns using feature scaling
WorkWeekHrs=WorkWeekHrs.dropna()
CodeRevHrs=CodeRevHrs.dropna()
norm_work=(WorkWeekHrs['WorkWeekHrs']-WorkWeekHrs['WorkWeekHrs'].min())/(WorkWeekHrs['WorkWeekHrs'].max()-WorkWeekHrs['WorkWeekHrs'].min())
norm_code=(CodeRevHrs['CodeRevHrs']-CodeRevHrs['CodeRevHrs'].min())/(CodeRevHrs['CodeRevHrs'].max()-CodeRevHrs['CodeRevHrs'].min())
ax0=WorkWeekHrs.plot(kind='scatter',x='Age',y='WorkWeekHrs',color='darkblue',
                 s=norm_work*2000+10,figsize=(15,7),alpha=0.30,ylim=(0,100))

ax1=CodeRevHrs.plot(kind='scatter',x='Age',y='CodeRevHrs',color='darkgreen',figsize=(15,7),
                 s=norm_code*2000+10,alpha=0.30,ax=ax0)

ax0.set_title('Bubble Plot of  WorkWeekHrs and CodeRevHrs')
ax0.legend(['WorkWeekHrs','CodeRevHrs'],loc='upper left')

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
Query='''
SELECT * FROM DatabaseDesireNextYear
'''
Data=pd.read_sql(Query,conn)
Database=Data.groupby(['DatabaseDesireNextYear']).sum()
Database=Database.sort_values(ascending=False,by='Respondent')
Top5_DBS=Database.head(5)
color_lib={'red','darkblue','darkgreen','yellow','orange'}
labels=['PostgreSql','MongoDB','Redis','MySQL','Elaticsearch']
Top5_DBS.plot(kind='pie',colors=color_lib,shadow=True,
              startangle=90, labels=labels,figsize=(10,7),autopct='%1.f%%',subplots=True)
plt.title('Pie Chart of The Top 5 Databases Desired to learn')
plt.legend(labels=labels,loc='upper left')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query='''
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM Master
'''
stack=pd.read_sql(Query,conn)

stack['Age']=stack['Age'].fillna(28) #fill missing values so Age can be used as index
stack['Age']=stack['Age'].astype(int)#convert to int to be used for index

stack['CodeRevHrs'].fillna(2.0) #fill missing values for CodeRevHrs by inputing the mode
filstack=stack[(stack['Age']>=30) & (stack['Age']<=35)]
filstackgrp=filstack.groupby(['Age'])[['WorkWeekHrs', 'CodeRevHrs']].median()
filstackgrp.head()
filstackgrp.plot(kind='area',alpha=0.35, figsize=(10,7),stacked=False)
plt.title('Median WorkWeekHrs and CodeRevHrs for Age group 30-35')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
Query='''
SELECT Age,ConvertedComp FROM
Master
'''
Concomp=pd.read_sql(Query,conn)
Concompfil=Concomp[(Concomp['Age']>=45)&(Concomp['Age']<=60)]
condata=Concompfil.groupby('Age').median()

condata.plot(kind='line',color='orange',figsize=(10,7))
plt.ylabel('Compensation')
plt.title('Converted Compensation For Ages From 45 to 60')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
Query='''
SELECT MainBranch FROM Master
'''
mainbranch=pd.read_sql(Query,conn)
#using dummy variables to attach numerical data to the options
dummy1=pd.get_dummies(mainbranch)
dummy1.rename(columns={'MainBranch_I am a developer by profession':'MainBranch_I am not primarily a developer, but I write code sometimes as part of my work': 'Not Primaly,Partly For Work'},inplace=True)
dumm=dummy1.sum()
dumm.plot(kind='barh',color='blue',figsize=(10,7))
plt.title('Bar Chart using the MainBranch column')
plt.show()
```
